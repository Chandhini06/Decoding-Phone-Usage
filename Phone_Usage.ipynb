```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# Standardize numerical features
numerical_cols = ['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)', 'Calls Duration (mins/day)',
                      'Number of Apps Installed', 'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)',
                      'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)', 'Monthly Recharge Cost (INR)']
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
X = df.drop("Primary Use", axis = 1)
y = df["Primary Use"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a Random Forest model to get feature importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Get feature importances
feature_importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)

# Print top features
print("Important Features:\n", feature_importances[:15])
```

```python
selected_features = feature_importances[feature_importances > 0.05].index  
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]
```

```python

# # Define hyperparameter grids
param_grids = {
    "XGBClassifier": {
        'n_estimators': [50, 100, 150],
        'max_depth': [3, 5, 7, 10],
        'learning_rate': [0.01,0.05, 0.1]
        # 'subsample': [0.8, 0.9, 1.0],  # Add subsample for regularization
        # 'colsample_bytree': [0.8, 0.9, 1.0],  # Add column sampling
        # 'gamma': [0, 0.1, 0.2],  # Regularization parameter
        # 'reg_alpha': [0, 0.1, 0.5],  # L1 regularization
        # 'reg_lambda': [0, 0.1, 0.5] 
        
    },
    "DecisionTreeClassifier": {
        'max_depth': [3, 5, 7, 10],
        'min_samples_split': [3, 5, 7, 10],
        'min_samples_leaf': [1, 2, 4, 5]
        # 'max_features': ['sqrt', 'log2']
    },
    "RandomForestClassifier": {
        'n_estimators': [50, 100, 150],
        'max_depth': [3, 5, 7, 10],
        'min_samples_split': [3, 5, 7, 10]
        # 'max_features': ['sqrt', 'log2']
    }
}

# Define models
models = {
    "LogisticRegression" : LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42),
    "XGBClassifier": XGBClassifier(random_state=42),
    "DecisionTreeClassifier": DecisionTreeClassifier(random_state=42),
    "RandomForestClassifier": RandomForestClassifier(random_state=42)
    # "GradientBoostingClassifier": GradientBoostingClassifier(random_state=42)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


# Store best models
best_models = {}

for model_name, model in models.items():
   
    if model_name == "LogisticRegression":
        print(f"Skipping hyperparameter tuning for {model_name} (no tuning required)")
        best_models[model_name] = model
    else:
        print(f"Tuning {model_name}...")
        search = RandomizedSearchCV(model, param_grids[model_name], cv=cv, n_iter=3, scoring='accuracy', n_jobs=-1,random_state=42, verbose=0)
        search.fit(X_train_selected, y_train)
        best_models[model_name] = search.best_estimator_

    # Get the best model
        best_models[model_name] = search.best_estimator_
        print(f" Best Params for {model_name}: {search.best_params_}\n")

# Evaluate models
for model_name, model in best_models.items():
    model.fit(X_train_selected, y_train)
    train_pred = model.predict(X_train_selected)
    test_pred = model.predict(X_test_selected)
    
    print (f"Model: {type(model).__name__}")

    
    print("*******Train*******")
    print(f"Accuracy Score: {accuracy_score(y_train, train_pred):.4f}")
    print(f"Precision Score: {precision_score(y_train, train_pred, average='macro'):.4f}")
    print(f"Recall Score: {recall_score(y_train, train_pred, average='macro'):.4f}")
    print(f"F1 Score: {f1_score(y_train, train_pred, average='macro'):.4f}")
    print("\n")
        
    print("********Test********")
    print(f"Accuracy Score: {accuracy_score(y_test, test_pred):.4f}")
    print(f"Precision Score: {precision_score(y_test, test_pred, average='macro'):.4f}")
    print(f"Recall Score: {recall_score(y_test, test_pred, average='macro'):.4f}")
    print(f"F1 Score: {f1_score(y_test, test_pred, average='macro'):.4f}")
    print("\n")
```

```python
import pickle
filename = f"{model_name.replace(' ', '_').lower()}_model.pkl"  
with open(filename, "wb") as f:
    pickle.dump(best_models, f)

    print(f" Saved {model_name} model as {filename}")
```

```python
import pickle
from sklearn.metrics import accuracy_score

# Track the best model
best_model_name = None
best_model = None
best_accuracy = 0

# Find the best model based on test accuracy
for model_name, model in best_models.items():
    test_pred = model.predict(X_test_selected)
    test_accuracy = accuracy_score(y_test, test_pred)
    
    print(f"Model: {model_name} - Test Accuracy: {test_accuracy:.4f}")
    
    if test_accuracy > best_accuracy:
        best_accuracy = test_accuracy
        best_model = model
        best_model_name = model_name

# Save the best model
filename = f"{best_model_name.lower()}_best_model.pkl"
with open(filename, "wb") as f:
    pickle.dump(best_model, f)

print(f"\n Best Model: {best_model_name} with Accuracy: {best_accuracy:.4f}")
print(f" Model saved as {filename}")
```

```python
from sklearn.cluster import KMeans, DBSCAN, SpectralClustering, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
clustering_results = {}
```

```python
# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_train_selected)
kmeans_silhouette = silhouette_score(X_train_selected, kmeans_labels)
clustering_results['KMeans'] = {'Model': kmeans,'Labels': kmeans_labels, 'Silhouette Score': kmeans_silhouette}
print(f'K-Means Silhouette Score: {kmeans_silhouette:.4f}')
```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.99, min_samples=4)
dbscan_labels = dbscan.fit_predict(X_train_selected)
dbscan_silhouette = silhouette_score(X_train_selected, dbscan_labels) if len(set(dbscan_labels)) > 1 else np.nan
clustering_results['DBSCAN'] = {'Model': dbscan,'Labels': dbscan_labels, 'Silhouette Score': dbscan_silhouette}
print(f'DBSCAN Silhouette Score: {dbscan_silhouette:.4f}')

```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Compute the k-nearest neighbors distances
k = 5  # min_samples value
neigh = NearestNeighbors(n_neighbors=k)
nbrs = neigh.fit(X_train_selected)
distances, indices = nbrs.kneighbors(X_train_selected)

# Sort distances and plot
distances = np.sort(distances[:, k-1])
plt.plot(distances)
plt.xlabel("Data Points")
plt.ylabel("k-Nearest Distance")
plt.title("Elbow Method for Optimal eps")
plt.show()
```

```python
# Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X_train_selected)
gmm_silhouette = silhouette_score(X_train_selected, gmm_labels)
clustering_results['GMM'] = {'Model': gmm,'Labels': gmm_labels, 'Silhouette Score': gmm_silhouette}
print(f'Gaussian Mixture Model Silhouette Score: {gmm_silhouette:.4f}')
```

```python
# Spectral Clustering
spectral = SpectralClustering(n_clusters=2, random_state=42, affinity='nearest_neighbors')
spectral_labels = spectral.fit_predict(X_train_selected)
spectral_silhouette = silhouette_score(X_train_selected, spectral_labels)
clustering_results['Spectral'] = {'Model': spectral,'Labels': spectral_labels, 'Silhouette Score': spectral_silhouette}
print(f'Spectral Clustering Silhouette Score: {spectral_silhouette:.4f}')
```

```python
# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(X_train_selected)
agglo_silhouette = silhouette_score(X_train_selected, agglo_labels)
clustering_results['Agglomerative'] = {'Model': agglo,'Labels': agglo_labels, 'Silhouette Score': agglo_silhouette}
print(f'Agglomerative Clustering Silhouette Score: {agglo_silhouette:.4f}')
```

```python
for method, result in clustering_results.items():
    print(f'\n{method} Clustering:')
    print(f'Silhouette Score: {result["Silhouette Score"]:.4f}')
```

```python
best_model_name = max(clustering_results, key=lambda k: clustering_results[k]['Silhouette Score'])
best_model = clustering_results[best_model_name]['Model']
print(f'Best Model: {best_model_name} with Silhouette Score: {clustering_results[best_model_name]["Silhouette Score"]:.4f}')

# Save the best model
with open(f'{best_model_name}_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)

print(f'{best_model_name} model saved as {best_model_name}_model.pkl')
```
